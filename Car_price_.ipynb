# **ML**

**ðŸ‘‰Making Dataset Copy**

```python
data2 = data.copy()
```

**ðŸ‘‰Label Encoding For categorical Data**

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
data2['drivewheel'] = label.fit_transform(data2['drivewheel'])
data2['carbody'] = label.fit_transform(data2['carbody'])
data2['cylindernumber'] = label.fit_transform(data2['cylindernumber'])
data2['fueltype'] = label.fit_transform(data2['fueltype'])
data2['aspiration'] = label.fit_transform(data2['aspiration'])
data2['doornumber'] = label.fit_transform(data2['doornumber'])
data2['enginelocation'] = label.fit_transform(data2['enginelocation'])
data2['enginetype'] = label.fit_transform(data2['enginetype'])
data2['fuelsystem'] = label.fit_transform(data2['fuelsystem'])
data2.head()
```

**ðŸ‘‰Droping unwanted columns** 

```python
cleandata = data2.drop(columns=["price","car_ID","CarName"])
```

```python
r = cleandata.iloc[ :, :-1].values
o = data2.iloc[:, -1].values
```

**ðŸ‘‰Splitting the dataset into the training set and test set**

```python
from sklearn.model_selection import train_test_split
r_train, r_test, o_train, o_test = train_test_split(r , o, test_size = 0.2, random_state = 1)
```

**ðŸ‘‰Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
r_train[:, :-1] = sc.fit_transform(r_train[ :, :-1])
r_test[:, :-1] = sc.transform(r_test[:, :-1])
```

**ðŸ‘‰Importing regression Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

**ðŸ‘‰Training regression Model on the Trainning set**

```python
models2 = [
    LinearRegression(),
    DecisionTreeRegressor(),
    RandomForestRegressor(),
    SVR(),
    KNeighborsRegressor(),
    MLPRegressor()
]
for model2 in models2:

    model2.fit(r_train, o_train)

    o_pred = model2.predict(r_test)

    r2 = round(r2_score(o_test, o_pred), 2)

    print(type(model2).__name__, "R2 score:", r2)
```
